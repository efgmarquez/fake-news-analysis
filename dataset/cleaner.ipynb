```python
df = pd.read_csv("data.csv")
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
df['Date posted'] = pd.to_datetime(df['Date posted']).dt.tz_convert('Asia/Manila')
df['Joined'] = pd.to_datetime(df['Joined']).dt.tz_convert('Asia/Manila')
```

```python
df
```

```python
df[['Group', 'Collector']] = ''
df['Category'] = 'ECON'
df['Group'] = '02'
df[['Timestamp','Tweet Translated', 'Screenshot', 'Content type', 'Tweet Type']] = ''
df['Topic'] = 'Marcos Sr. Wealth to Filipinos'
df = df.rename({'Date posted': 'Date Posted'}, axis=1)
df['Account handle'] = df['Account handle'].map(lambda x: '@'+x)
```
